```python
def calculate_energy_consumption(file_path, tile_name):
    # Load the Excel file
    xls = pd.ExcelFile(file_path)
    
    # Display sheet names to understand the structure
    sheet_names = xls.sheet_names
    
    # Load the data from the first sheet
    df = xls.parse(sheet_names[0])

    # Display the first few rows to understand the structure
    df.head()
    
    # Extract active power values
    active_power = df['peak_value'].values

    # Create a time axis (in hours), assuming 288 samples evenly spaced over 24 hours
    time_hours = np.linspace(0, 24, len(active_power))

    # Plot the active power
    plt.figure(figsize=(12, 6))
    plt.plot(time_hours, active_power, label='Active Power (W)', color='blue')
    plt.xlabel('Time (Hours)')
    plt.ylabel('Active Power (W)')
    plt.title(tile_name)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Calculate area under the curve using trapezoidal rule
    area_under_curve = np.trapezoid(active_power, time_hours)

    print("Totale kWh: ",area_under_curve*0.001, "\n")  # Convert from Wh to kWh
    print("Total kWh: {:.2f} kWh".format(area_under_curve * 0.001))  # Print total kWh with two decimal place
    
```

```python
file_path = r"C:\Users\hp\Desktop\REG108S\Project\MatLab Data\power 20250506 110500 to 20250508 110000\Total Loads for 7Mayto8May.xlsx"
calculate_energy_consumption(file_path, "Total Active Power form 7 May to 8 May")
```
