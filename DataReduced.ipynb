```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Excel data
# Change 'your_data.xlsx' and 'Sheet1' if needed
df = pd.read_excel(r"c:\Users\hp\Desktop\REG108S\Project\MatLab Data\power 20250503 to 20250504\ThreePhaseLoads.xlsx")

# Replace 'value' with the actual column name in your Excel file
data = df['Active Power'].values

# Check for 288 points
assert len(data) == 289, "Input data must contain exactly 288 points."

# Store hourly peaks
hourly_peaks = []

for i in range(24):
    chunk = data[i*12:(i+1)*12]
    peak = np.max(chunk)
    hourly_peaks.append(peak)

# Save to Excel
peak_df = pd.DataFrame({'hour': [f"{h:02}:00" for h in range(24)],
                        'peak_value': hourly_peaks})
peak_df.to_excel('3phaseLoadPeak.xlsx', index=False)

# Plot original vs peaks
plt.figure(figsize=(14, 5))
plt.plot(data, label='Original (288 pts)', alpha=0.4)
plt.plot(np.linspace(0, 288, 24), hourly_peaks, 'ro-', label='Hourly Peaks (24 pts)')
plt.title('Hourly Peak Points Extracted from Time Series')
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## New code for data reduction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

# Load Excel data
file_path = r"C:\Users\hp\Desktop\REG108S\Project\MatLab Data\power 20250507 110500 to 20250508 110000\individual_load_actual_power_profiles.xlsx"
df = pd.read_excel(file_path)

# Replace with actual column name
data = df['peak_value'].values
total_points = len(data)

# Ensure enough data to create 24 bins
if total_points < 24:
    raise ValueError("Input data must have at least 24 points")

# Divide into 24 chunks
samples_per_hour = total_points // 24
reduced_data = []

for i in range(24):
    start = i * samples_per_hour
    end = (i + 1) * samples_per_hour if i < 23 else total_points  # Last chunk may include remainder
    chunk = data[start:end]
    peak = np.max(chunk)  # or use np.mean(chunk) if average is preferred
    reduced_data.append(peak)

# Create DataFrame
peak_df = pd.DataFrame({
    'hour': [f"{h:02}:00" for h in range(24)],
    'peak_value': reduced_data
})
peak_df.to_excel('Total Loads for 7Mayto8May.xlsx', index=False)

# Interpolate reduced data back to original length for error comparison
interpolated = np.interp(np.linspace(0, 23, total_points), np.arange(24), reduced_data)

# Calculate % error using MAPE
mape = mean_absolute_percentage_error(data, interpolated) * 100

# Plot original vs reduced
plt.figure(figsize=(14, 5))
plt.plot(data, label='Original ({} pts)'.format(total_points), alpha=0.5)
plt.plot(np.linspace(0, total_points - 1, 24), reduced_data, 'ro-', label='Reduced (24 pts)')
plt.title(f'Original vs Reduced Time Series for 7May to 8May (MAPE: {mape:.2f}%)')
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Percentage Error (MAPE): {mape:.2f}%")
```

## Another code for data reduction

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

def downsample_and_evaluate(data, target_points=24, method='mean'):
    n = len(data)
    segment_size = n // target_points
    reduced = []

    for i in range(target_points):
        start = i * segment_size
        end = (i + 1) * segment_size if i < target_points - 1 else n
        segment = data[start:end]

        if method == 'mean':
            reduced.append(np.mean(segment))
        elif method == 'max':
            reduced.append(np.max(segment))
        elif method == 'min':
            reduced.append(np.min(segment))
        else:
            raise ValueError("Unsupported method")

    # Interpolate to original length for error computation
    interpolated = np.interp(np.linspace(0, target_points - 1, n),
                             np.arange(target_points), reduced)
    mape = mean_absolute_percentage_error(data, interpolated) * 100
    return np.array(reduced), mape
```

```python
import pandas as pd

# Load data
file_path = r"C:\Users\hp\Desktop\REG108S\Project\MatLab Data\power 20250503 to 20250504\individual_load_actual_power_profiles.xlsx"
df = pd.read_excel(file_path)
data = df['peak_value'].values

# Try all three
methods = ['mean', 'max', 'min']
results = {}

for method in methods:
    reduced, mape = downsample_and_evaluate(data, 24, method=method)
    results[method] = (reduced, mape)
    print(f"{method.upper()} MAPE: {mape:.2f}%")

# Plot comparison
plt.figure(figsize=(12, 6))
plt.plot(data, label='Original', alpha=0.4)
for method in methods:
    plt.plot(np.linspace(0, len(data)-1, 24), results[method][0], label=f'{method} (MAPE: {results[method][1]:.2f}%)')
plt.legend()
plt.title("Downsampling Comparison")
plt.xlabel("Time Index")
plt.ylabel("Power Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Another code 2

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_percentage_error
from scipy.interpolate import splrep, splev

def spline_downsample_and_evaluate(data, target_points=24):
    n = len(data)
    x_original = np.linspace(0, 1, n)  # normalize time axis
    x_target = np.linspace(0, 1, target_points)

    # Fit B-spline to the original signal
    tck = splrep(x_original, data, s=0)  # s=0 => no smoothing, exact fit
    reduced = splev(x_target, tck)

    # Interpolate back to original length for error comparison
    interpolated = splev(x_original, tck)
    mape = mean_absolute_percentage_error(data, interpolated) * 100

    return np.array(reduced), mape
```

```python
# Load your Excel file
file_path = r"C:\Users\hp\Desktop\REG108S\Project\MatLab Data\power 20250503 to 20250504\individual_load_actual_power_profiles.xlsx"
df = pd.read_excel(file_path)
data = df['peak_value'].values

# Run spline downsampling
reduced_spline, mape_spline = spline_downsample_and_evaluate(data, 24)

# Print result
print(f"Spline Fit MAPE: {mape_spline:.2f}%")

# Plot
plt.figure(figsize=(12, 6))
plt.plot(data, label='Original', alpha=0.4)
plt.plot(np.linspace(0, len(data) - 1, 24), reduced_spline, 'r.-', label=f'Spline Downsampled (MAPE: {mape_spline:.2f}%)')
plt.legend()
plt.xlabel('Time Index')
plt.ylabel('Value')
plt.title('Spline Downsampling vs Original')
plt.grid(True)
plt.tight_layout()
plt.show()
```
